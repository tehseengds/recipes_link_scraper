```python
import glob
import json

# Path to the folder containing the JSON files
folder_path = 'links/'

# Get a list of all JSON files in the folder
json_files = glob.glob(folder_path + '*.json')

#total length of links in all files
total_length = 0

# Read the contents of each JSON file and print their lengths
for file_path in json_files:
    with open(file_path, 'r') as file:
        data = json.load(file)
        file_length = len(data)
        total_length += file_length
        print(f"File: {file_path}, Length: {file_length}")

print(f"Total length of links: {total_length}")
```
