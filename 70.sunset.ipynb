```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, StaleElementReferenceException
```

```python
# Set up the webdriver (make sure you have the appropriate webdriver installed)
driver = webdriver.Chrome()


# Navigate to the website
url = "https://www.sunset.com/recipes"

driver.get(url)
```

```python
import json
import time

links = []
i = 1

while True:
    try:
        articles = driver.find_elements(By.XPATH, '//section[@class="cooked-clearfix cooked-recipe-grid cooked-recipe-loader cooked-columns-3"]/article[@class="cooked-recipe has-post-thumbnail cooked-recipe-card cooked-recipe-card-modern-centered"]/a')

        for article in articles:
            link = article.get_attribute("href")
            links.append(link)
            print(link)


        next_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//a[@class="next page-numbers"]')))


        next_button.click()
        print(f"Page Completed {i}")
        i += 1

        time.sleep(3)


    except Exception as e:
        print(e)
        break


# Remove duplicates from the links list
links = list(set(links))

# Define the filepath for the JSON file
file_name = "./links/70.sunset2.json"

# Open the file & Write data to JSON file
with open(file_name, "w") as f:
    json.dump(links, f, indent=4)
```
