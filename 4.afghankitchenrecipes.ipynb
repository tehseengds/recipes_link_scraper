```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
```

```python
# Set up the webdriver (make sure you have the appropriate webdriver installed)
driver = webdriver.Chrome()

# Navigate to the website
url = "http://www.afghankitchenrecipes.com/recent-recipes/"
driver.get(url)
```

```python
import json

# Create a list to store the links
links = []

while True:
    try:
        # Find all the anchar tag elements
        articles = driver.find_elements(By.XPATH, '//*[@class="post-thumb single-img-box"]/a')

        for article in articles:

            # Get the href attribute (link)
            link = article.get_attribute("href")

            # Append the link to the list
            links.append(link)

        # Find the "Next" button
        next_button = WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.LINK_TEXT, 'Next Â»')))

        # Click the "Next" button
        next_button.click()

    except Exception as e:
        # Break the loop if there's no "Next" button or any other exception occurs
        print(f"An error occurred: {e}")
        break

# Remove duplicates from the links list
links = list(set(links))

# Define the filepath for the JSON file
file_name = "./links/4.afghankitchenrecipes.json"

# Save the links to the JSON file
with open(file_name, 'w') as file:
    json.dump(links, file)
```
