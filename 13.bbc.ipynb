```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
```

```python
# Set up the webdriver (make sure you have the appropriate webdriver installed)
driver = webdriver.Chrome()

# Navigate to the website
url = "https://www.bbc.co.uk/food/recipes/a-z/"
driver.get(url)
```

```python
# # # jsut for A # # #

import json

links = []
i = 1


while True:
    try:
        grid = driver.find_element(By.XPATH, '//*[@class="gel-layout gel-layout--equal promo-collection standard-12-promos"]/div/a')
        articles = driver.find_elements(By.XPATH, '//*[@class="gel-layout gel-layout--equal promo-collection standard-12-promos"]/div/a')

        for article in articles:
            link = article.get_attribute("href")
            links.append(link)
            print(link)

        # Find all list items
        next_buttons = driver.find_elements(By.XPATH, '//ul[@class="pagination__list"]/li')

        # Select the last item from the list
        last_button = next_buttons[-1]

        driver.execute_script("arguments[0].click();", last_button)
        print(f"Clicked Next {i}")
        i += 1

        

    except Exception as e:
        print(e)
        break


# Remove duplicates from the links list
# links = list(set(links))

# Define the filepath for the JSON file
file_name = "./links/a.json"

# Open the file & Write data to JSON file
with open(file_name, "w") as f:
    json.dump(links, f, indent=4)
```
