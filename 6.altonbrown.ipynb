```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
```

```python
# Set up the webdriver (make sure you have the appropriate webdriver installed)
driver = webdriver.Chrome()

# Navigate to the website
url = "https://altonbrown.com/cook/"
driver.get(url)
```

```python
while True:
    try:
        button = driver.find_element(By.XPATH, '//*[@id="ajax-load-more"]/div[2]/button')
        driver.execute_script("arguments[0].click();", button)
        print("Clicked")
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    except Exception as e:
        print(e)
        break
```

```python
import json

links = []

try:
    # Get all the articles on the page
    articles = driver.find_elements(By.XPATH, '//div[@class="recipe-item"]/a')
    
    for article in articles:
        # Get the link to the article
        link = article.get_attribute("href")

        # Add the link to the list of links
        links.append(link)
        print(link)
    
except Exception as e:
    print(e)

# Remove duplicates from the links list
links = list(set(links))

# Define the filepath for the JSON file
file_name = "./links/6.altonbrown.json"

# Open the file & Write data to JSON file
with open(file_name, "w") as f:
    json.dump(links, f, indent=4)
```
